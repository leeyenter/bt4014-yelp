```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
review = pd.read_json('data/review.json', lines=True, orient='columns')
business = pd.read_json('data/business.json', lines=True, orient='columns')
```

```python
review.head()
```

```python
business.set_index('business_id', inplace=True)
```

```python
business.head()
```

```python
business_size = business.groupby('name').size()
```

```python
business_size = pd.DataFrame(business_size)
business_size.columns = ['num_outlets']
```

```python
business_size.shape
```

```python
business_size.sort_values('num_outlets', ascending=False).head(20)
```

```python
review_size = review.groupby('business_id').size()
review_size = pd.DataFrame(review_size)
review_size.columns = ['num_reviews']
```

```python
review_size.head()
```

```python
df = pd.DataFrame(business.groupby([business.index.get_level_values(0), 'name']).size()).reset_index().drop(0, axis=1)
```

```python
df = df.set_index('business_id').join(review_size).groupby('name').sum()
```

```python
df = df.join(business_size)
```

```python
fig, ax = plt.subplots(figsize=(16, 8))
df.num_reviews.hist(bins=20)
plt.yscale('log')
plt.ylabel('No. of businesses')
plt.xlabel('No. of reviews')
```

```python
fig, ax = plt.subplots(figsize=(16, 8))
df.num_outlets.hist(bins=20)
plt.yscale('log')
plt.ylabel('No. of businesses')
plt.xlabel('No. of outlets')
```

```python
fig, ax = plt.subplots(figsize=(16, 8))
sns.scatterplot(x = df.num_outlets, y = df.num_reviews)
# plt.yscale('log')
# plt.xscale('log')
# plt.ylim(1, df.num_reviews.max())
# plt.xlim(1, df.num_outlets.max())
```
